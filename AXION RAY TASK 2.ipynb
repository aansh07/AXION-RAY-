## 4. Text Analysis for Tag Generation

```python
def extract_tags(text):
    text = str(text).lower()
    tags = []
    
    # Common failure modes
    failure_modes = {
        'leak': ['leak', 'dripping', 'seep'],
        'electrical': ['wire', 'harness', 'sensor', 'module', 'circuit'],
        'mechanical': ['loose', 'tighten', 'bolt', 'nut', 'clip'],
        'material': ['peel', 'stitch', 'crack', 'bubble', 'fray']
    }
    
    for mode, keywords in failure_modes.items():
        if any(keyword in text for keyword in keywords):
            tags.append(mode)
    
    # Common components
    components = {
        'steering_wheel': ['steering wheel', 'wheel'],
        'module': ['module', 'control module'],
        'harness': ['harness', 'wiring'],
        'trim': ['trim', 'cover', 'bezel']
    }
    
    for comp, keywords in components.items():
        if any(keyword in text for keyword in keywords):
            tags.append(comp)
    
    return ', '.join(tags) if tags else 'other'

df['TAGS'] = df['CUSTOMER_VERBATIM'].apply(extract_tags) + '; ' + df['CORRECTION_VERBATIM'].apply(extract_tags)
```
